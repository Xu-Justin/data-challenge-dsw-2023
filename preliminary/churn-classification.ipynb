```python
import pandas as pd

CSV_FILE = 'Telco_customer_churn_adapted_v2.xlsx'

df = pd.read_excel(CSV_FILE)
df.head()
```

```python
encode_columns = ['Location', 'Device Class', 'Payment Method']
encode_yn_columns = ['Games Product', 'Music Product', 'Education Product', 'Call Center', 'Video Product', 'Use MyApp', 'Churn Label']

encoded_df = pd.get_dummies(df, columns=encode_columns)
for column in encode_yn_columns:
    encoded_df[column] = encoded_df[column] == 'Yes'

encoded_df.info()
encoded_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scale_columns = ['Tenure Months', 'Monthly Purchase (Thou. IDR)', 'CLTV (Predicted Thou. IDR)']

scaler = StandardScaler()
scaled_df = encoded_df.copy()
scaled_df[scale_columns] = scaler.fit_transform(encoded_df[scale_columns])

scaled_df.info()
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split

non_feature_columns = ['Customer ID', 'Longitude', 'Latitude', 'Churn Label']
target_column = 'Churn Label'

feature = scaled_df.drop(non_feature_columns, axis=1)
target = scaled_df[target_column]

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=0)
print('Shape of training feature:', X_train.shape)
print('Shape of testing feature :', X_test.shape)
print('Shape of training label  :', y_train.shape)
print('Shape of testing label   :', y_test.shape)

X_train.head()
```

```python
from sklearn import metrics

def evaluate_model(model, x_test, y_test):
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)
    kappa = metrics.cohen_kappa_score(y_test, y_pred)

    # Calculate area under curve (AUC)
    y_pred_proba = model.predict_proba(x_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    
    return {
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1-score': f1,
        'cohen-kappa': kappa,
#         'fpr': fpr,
#         'tpr': tpr,
        'area-under-curve': auc,
        'true-positive': cm[0, 0],
        'false-positive': cm[0, 1],
        'false-negative': cm[1, 0],
        'true-negative': cm[1, 1],
    }

def print_metrics(model_metrics):
    for key, value in model_metrics.items():
        print(f'{key:20s}: {value}')
```

```python
from sklearn import tree

dtc = tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
dtc_eval = evaluate_model(dtc, X_test, y_test)

print_metrics(dtc_eval)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
rf_eval = evaluate_model(rf, X_test, y_test)

print_metrics(rf_eval)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)
nb_eval = evaluate_model(nb, X_test, y_test)

print_metrics(nb_eval)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_eval = evaluate_model(knn, X_test, y_test)

print_metrics(knn_eval)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [50, 80, 100],
    'max_features': [2, 3, 4],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 300, 500, 750, 1000]
}

rf_grids = RandomForestClassifier(random_state=0)
grid_search = GridSearchCV(estimator=rf_grids, param_grid=param_grid, scoring='recall', cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
best_grid = grid_search.best_estimator_
best_grid_eval = evaluate_model(best_grid, X_test, y_test)

print_metrics(best_grid_eval)
```
