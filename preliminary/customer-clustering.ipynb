```python
import pandas as pd

CSV_FILE = 'Telco_customer_churn_adapted_v2.xlsx'

df = pd.read_excel(CSV_FILE)
df.head()
```

```python
encode_columns = ['Location', 'Device Class', 'Payment Method']
encode_yn_columns = ['Games Product', 'Music Product', 'Education Product', 'Call Center', 'Video Product', 'Use MyApp', 'Churn Label']

encoded_df = pd.get_dummies(df, columns=encode_columns)
for column in encode_yn_columns:
    encoded_df[column] = encoded_df[column] == 'Yes'

encoded_df.info()
encoded_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scale_columns = ['Tenure Months', 'Monthly Purchase (Thou. IDR)']

scaler = StandardScaler()
scaled_df = encoded_df.copy()
scaled_df[scale_columns] = scaler.fit_transform(encoded_df[scale_columns])

scaled_df.info()
scaled_df.head()
```

```python
from sklearn.feature_selection import VarianceThreshold

# Drop Non Feature Columns
non_features_columns = ['Customer ID', 'Longitude', 'Latitude', 'Churn Label', 'CLTV (Predicted Thou. IDR)']
features = scaled_df.drop(non_features_columns, axis=1)

# Variable Selection using Variance Threshold
threshold = 0.2
selector = VarianceThreshold(threshold)
selector.fit(features)
selected_column_names = features.columns[selector.get_support(indices=True)]
features = features[selected_column_names]
features.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.decomposition import KernelPCA

n_components = 3
features_decompose = features.copy()


pca = PCA(n_components=n_components)
res_pca = pca.fit_transform(features)
columns_pca = [f'pca_{i}' for i in range (n_components)]
features_decompose[columns_pca] = res_pca


kpca_poly = KernelPCA(n_components=n_components, kernel="poly", degree=3)
res_kpca_poly = kpca_poly.fit_transform(features)
columns_kpca_poly = [f'kpca_poly_{i}' for i in range (n_components)]
features_decompose[columns_kpca_poly] = res_kpca_poly


kpca_rbf = KernelPCA(n_components=n_components, kernel='rbf', gamma=0.01)
res_kpca_rbf = kpca_rbf.fit_transform(features)
columns_kpca_rbf = [f'kpca_rbf_{i}' for i in range (n_components)]
features_decompose[columns_kpca_rbf] = res_kpca_rbf


kpca_cosine = KernelPCA(n_components=n_components, kernel='cosine')
res_kpca_cosine = kpca_cosine.fit_transform(features)
columns_kpca_cosine = [f'kpca_cosine_{i}' for i in range (n_components)]
features_decompose[columns_kpca_cosine] = res_kpca_cosine


features_decompose.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn_extra.cluster import KMedoids
from sklearn.impute import SimpleImputer
from sklearn.metrics import silhouette_score
from cdbw import CDbw

def cluster(data, columns, model, k):
    imputer = SimpleImputer(strategy='median')
    X = imputer.fit_transform(data[columns])
    
    if model == 'KMeans':
        mdl = KMeans(n_clusters=k, random_state=0)
    elif model == 'KMedoids':
        mdl = KMedoids(n_clusters=k, random_state=0)
    elif model == 'Agglomerative':
        mdl = AgglomerativeClustering(n_clusters=k)
    else:
        raise ValueError(f'Unknown model: {model}')

    labels = mdl.fit_predict(X)
    score_silhouette = silhouette_score(X, labels)
    score_cdbw = CDbw(X, labels, metric="euclidean", alg_noise='comb', intra_dens_inf=False, s=3, multipliers=False)

    return score_silhouette, score_cdbw, labels

def compute_cluster(data, method, model, max_k=9):
    if method == 'PCA':
        columns = columns_pca
    elif method == 'KPCA Poly':
        columns = columns_kpca_poly
    elif method == 'KPCA RBF':
        columns = columns_kpca_rbf
    elif method == 'KPCA Cosine':
        columns = columns_kpca_cosine
    else:
        raise ValueError(f'Unknown method: {method}')
    
    max_score_silhouette = -1
    max_score_cdbw = -1
    max_k_silhouette = -1
    max_k_cdbw = -1
    
    for k in range(2, max_k+1):
        score_silhouette, score_cdbw, _ = cluster(data, columns, model, k)
        print(f'Cluster {method} data with {model} K={k}, silhouette score: {score_silhouette:8.6f}, cdbw_score: {score_cdbw:8.6f}')

        if score_silhouette > max_score_silhouette:
            max_score_silhouette = score_silhouette
            max_k_silhouette = k

        if score_cdbw > max_score_cdbw:
            max_score_cdbw = score_cdbw
            max_k_cdbw = k
            
    print(f'Best cluster {method} data with {model} based on Silhouette score, is K = {max_k_silhouette} with score: {max_score_silhouette:.6f}')
    print(f'Best cluster {method} data with {model} based on CDbw score, is K = {max_k_cdbw} with score: {max_score_cdbw:.6f}')
    print()
```

```python
compute_cluster(features_decompose, 'PCA', 'KMeans')
compute_cluster(features_decompose, 'KPCA Poly', 'KMeans')
compute_cluster(features_decompose, 'KPCA RBF', 'KMeans')
compute_cluster(features_decompose, 'KPCA Cosine', 'KMeans')
```

```python
compute_cluster(features_decompose, 'PCA', 'KMedoids')
compute_cluster(features_decompose, 'KPCA Poly', 'KMedoids')
compute_cluster(features_decompose, 'KPCA RBF', 'KMedoids')
compute_cluster(features_decompose, 'KPCA Cosine', 'KMedoids')
```

```python
compute_cluster(features_decompose, 'PCA', 'Agglomerative')
compute_cluster(features_decompose, 'KPCA Poly', 'Agglomerative')
compute_cluster(features_decompose, 'KPCA RBF', 'Agglomerative')
compute_cluster(features_decompose, 'KPCA Cosine', 'Agglomerative')
```

```python
def describe_cluster(data, data_decompose, columns, model, k):
    score_silhouette, score_cdbw, labels = cluster(data_decompose, columns, model, k)
    print(f'\n{model} K={k}, silhouette score: {score_silhouette:8.6f}, cdbw_score: {score_cdbw:8.6f}')
    
    describe_data = data.copy()
    describe_data['labels'] = labels
    for i in range(k):
        print(f'\n{"-" * 30}\n\n> Describe for cluster {i}\n')
        print(describe_data[describe_data['labels'] == i].describe())

describe_cluster(df, features_decompose, columns_kpca_rbf, 'KMeans', 7)
```
